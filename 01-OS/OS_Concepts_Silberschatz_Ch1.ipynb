# Ch1 Introduction

#### 作業系統是一種管理電腦硬體的軟體。它同時也為應用程式提供基礎，並充當電腦使用者與電腦硬體之間的中介。作業系統最令人驚奇的一點，在於它們在各種運算環境中完成這些任務的方式各有不同。從汽車與家庭電器（包含物聯網裝置）、智慧型手機、個人電腦、企業級電腦，到雲端運算環境，作業系統無所不在。

#### 為了探索作業系統在現代運算環境中的角色，我們必須先了解電腦硬體的組織與架構。這包括中央處理器（CPU）、記憶體、輸入/輸出（I/O）裝置，以及儲存裝置。作業系統的一項基本職責，就是將這些資源分配給各個程式使用。

#### 由於作業系統龐大且複雜，它必須逐步建構而成。每個部分都應該是系統中界線清晰的模組，擁有明確定義的輸入、輸出與功能。在本章中，我們將概述當代電腦系統的主要組成部分，以及作業系統所提供的功能。此外，我們也將探討數個主題，以為後續章節鋪路：作業系統中使用的資料結構、運算環境，以及開放原始碼與免費作業系統。

### 章節目標
> • 說明電腦系統的一般組織方式與中斷的角色。  
> • 說明現代多處理器電腦系統的組成。  
> • 說明從使用者模式轉換至核心模式的過程。  
> • 討論作業系統如何應用於各種運算環境中。  
> • 提供免費與開放原始碼作業系統的實例。

### 1.1 作業系統的功能

我們的討論首先從作業系統在整體電腦系統中所扮演的角色開始。電腦系統大致可以劃分為四個組成部分：**硬體**、**作業系統**、**應用程式**，以及**使用者**（如圖 1.1 所示）。

硬體——包含中央處理器（CPU）、記憶體，以及輸入/輸出（I/O）裝置——提供了系統的基本運算資源。應用程式（例如文書處理器、試算表、編譯器與網頁瀏覽器）——定義了使用者如何利用這些資源來解決運算問題。作業系統則控制硬體並在不同應用程式及使用者之間協調其使用。

我們也可以將電腦系統看作由**硬體**、**軟體**與**資料**組成。作業系統提供恰當的機制，使得這些資源在整個電腦操作過程中得以被正確運用。作業系統就像一個政府；它本身並不產生實際效益，但建立了一個讓其他程式能夠有效運作的環境。

為了更深入理解作業系統的角色，我們將從「使用者觀點」與「系統觀點」來探討作業系統。

---

### 1.1.1 使用者觀點

使用者如何看待電腦系統，取決於其所使用的介面。許多使用者使用筆電或桌上型電腦，這些電腦配備螢幕、鍵盤與滑鼠，通常設計為單一使用者能完全掌控資源的情境。在這類系統中，作業系統的設計重點在於**使用便利性**，其次才是效能與安全性，而**資源使用效率**（例如資源共享的機制）則較不受重視。

如今，許多使用者轉而與行動裝置（例如智慧型手機與平板電腦）互動，這些裝置正逐步取代桌機與筆電的地位。這些設備通常透過行動網路或其他無線技術連接網路，並透過觸控螢幕與系統互動，用戶以手指滑動與點擊來操作，而不是使用實體鍵盤與滑鼠。許多行動裝置也支援語音操作，例如 Apple 的 Siri。

另一方面，有些電腦幾乎沒有使用者介面，例如嵌入式於家電或汽車中的電腦。這些裝置可能只有數字鍵盤或狀態指示燈，設計上主要是為了在**無人干預的情況下自動運行**。

---

### 1.1.2 系統觀點

從電腦本身的觀點來看，作業系統是**最密切操作硬體的程式**。在此脈絡中，我們可以將作業系統視為一種資源分配者。一台電腦擁有多種資源，包括 CPU 時間、記憶體空間、儲存空間與 I/O 裝置等。面對來自不同使用者與程式的資源需求，作業系統必須做出資源分配的決策，以確保電腦系統運作的**效率與公平性**。

另一種觀點是將作業系統視為**控制程式**。控制程式的職責是管理使用者程式的執行，避免錯誤與不當使用電腦資源，尤其是與 I/O 裝置的操作與控制密切相關。

---

### 1.1.3 作業系統的定義

此時，我們可以發現「作業系統」一詞涵蓋了**多種角色與功能**。這部分是因為電腦的設計與用途極其多樣。如今電腦存在於烤麵包機、汽車、船艦、太空船、家庭與商業場所，是遊戲機、有線電視機上盒與工業控制系統的基礎。

這種多樣性可從電腦的歷史解釋。雖然電腦歷史並不久遠，但其發展極為迅速。最初電腦是為了解決特定任務（如破譯密碼與彈道計算）而設計的固定用途機器，之後逐步演化為多功能大型主機系統，而作業系統也在這個階段誕生。1960 年代，摩爾定律預測積體電路上的電晶體數量每 18 個月將倍增，此預測至今大致仍準確。隨著功能的提升與體積的縮小，電腦用途與作業系統的種類也大幅增長。（更多歷史背景詳見附錄 A）

那麼，我們該如何定義作業系統呢？事實上，我們並沒有一個完全充分的定義。作業系統的存在，正是為了解決讓電腦可用這個問題。電腦系統的根本目標是**執行程式並幫助使用者解決問題**。硬體的設計也是為了這個目的。不過，僅有硬體並不容易操作，因此才會發展出應用程式。而這些程式往往需要共通功能（例如對 I/O 裝置的控制），這些共通功能就被整合成一個軟體——作業系統。

同樣地，對於作業系統的範圍，我們也沒有一致的看法。簡單的觀點是：**當你向廠商購買「作業系統」時，隨附的所有內容都算是作業系統**。然而，各系統間所附功能差異極大：有些系統僅佔用不到 1 MB 空間，甚至不包含全螢幕編輯器；另一些系統則需數 GB，並完全基於視窗化圖形介面。

較常見的定義，也是本書所採用的，是將作業系統視為**電腦中永遠在運行的程式**——通常稱為核心（kernel）。此外還有兩類程式：系統程式（與作業系統相關，但不屬於核心）與應用程式（所有與系統操作無關的程式）。

隨著個人電腦的普及與作業系統的複雜度提升，作業系統包含哪些功能也成為重要議題。1998 年，美國司法部控告微軟在作業系統中加入過多功能（例如內建瀏覽器），限制了應用程式供應商的競爭空間。法院最終判決微軟利用其作業系統壟斷地位來限制市場競爭。

然而，當我們觀察現代的行動裝置作業系統，會發現作業系統的功能再度擴張。行動作業系統通常不只包含核心，還包含「中介軟體（middleware）」——提供應用程式開發者額外服務的軟體架構。例如，Apple 的 iOS 與 Google 的 Android 系統都包括核心與支援資料庫、多媒體、圖形等的中介軟體。

總結來說，對我們而言，作業系統包含：**永遠在運行的核心程式**、**為應用開發提供支援的中介軟體框架**、以及**協助系統管理的系統程式**。本書大多聚焦於通用型作業系統的核心，但在需要時也會介紹其他元件，以全面說明作業系統的設計與運作。

> ### 為什麼要學習作業系統？

>  雖然有許多電腦科學從業人員，但其中只有極少數會參與作業系統的設計或修改。那麼，為什麼我們還要學習作業系統及其運作方式呢？原因很簡單：幾乎所有程式碼都是在作業系統之上執行的，因此，理解作業系統的運作原理對於正確、高效、有效且安全的程式設計至關重要。
>  
>  理解作業系統的基本原理、它們如何驅動電腦硬體，以及它們為應用程式提供了哪些功能，不僅對於開發作業系統的人來說是必要的，對於在其上撰寫與使用應用程式的人來說也同樣非常有幫助。
>  

### 1.2 電腦系統的組織

#### 一套現代通用型電腦系統由一個或多個中央處理器（CPU）與若干裝置控制器所組成，這些控制器透過一條共用的匯流排（bus）連接，提供元件間與共用記憶體之間的資料存取通道（如圖 1.2 所示）。每個裝置控制器負責一種特定的裝置類型（例如磁碟機、音訊裝置或圖形顯示器）。根據控制器的不同，有時一個控制器可以連接多個裝置。例如，一個系統的 USB 埠可以連接至 USB 集線器，而集線器又可連接多個裝置。裝置控制器會維持一些本地緩衝區以及一組特殊用途的暫存器，並負責控制它所連接的周邊裝置與緩衝區之間的資料傳輸。

#### 通常，作業系統會為每個裝置控制器配置一個**裝置驅動程式（device driver）**。此驅動程式了解該控制器的運作，並提供一個統一的介面給作業系統的其他部分。CPU 與裝置控制器可並行執行，彼此爭用記憶體資源。為了確保共用記憶體的存取秩序，需透過記憶體控制器來同步這些存取行為。

#### 接下來的子章節中，我們將說明這類系統運作的基本原理，聚焦於三個主要面向：**中斷機制（interrupts）**、**儲存結構（storage structure）**與**輸入/輸出結構（I/O structure）**。


### 1.2.1 中斷（Interrupts）

讓我們從一個典型的電腦操作範例開始：一個正在執行 I/O 操作的程式。為了啟動 I/O 操作，裝置驅動程式會將適當的資料寫入裝置控制器的暫存器。接著，裝置控制器會檢查這些暫存器的內容，以決定要採取的動作（例如「從鍵盤讀取一個字元」）。然後控制器開始將資料從裝置傳送至其本地緩衝區。

一旦資料傳輸完成，裝置控制器便會通知驅動程式其操作已結束。裝置驅動程式隨後會將控制權交還給作業系統的其他部分，若是讀取操作，可能會傳回資料或其指標；若是寫入操作，則會回報「寫入成功完成」或「裝置忙碌」等狀態訊息。但裝置控制器究竟是如何通知驅動程式其操作已完成的呢？這正是透過\*\*中斷（interrupt）\*\*來實現的。

---

### 1.2.1.1 概述（Overview）

硬體隨時都能透過系統匯流排向 CPU 發出訊號以觸發中斷。（儘管一套電腦系統可能有多條匯流排，但**系統匯流排**是各主要元件間的主要通訊路徑。）中斷的用途不僅止於 I/O，還是作業系統與硬體互動的關鍵機制。

當 CPU 接收到中斷時，它會立即中止當前的工作，轉而執行預設位置中的程式碼。這個固定位置儲存的是中斷服務程式（ISR, Interrupt Service Routine）的起始位址。該服務程式完成後，CPU 會繼續執行先前被中斷的任務。這一操作的時間流程如圖 1.3 所示。

中斷是電腦架構中不可或缺的一部分。雖然每種電腦設計皆有其專屬的中斷機制，但其中幾項功能是共通的：

* 中斷機制必須能夠將控制權轉移至對應的中斷服務程式。
* 傳統方式是使用通用例行程序來分析中斷資訊，並呼叫正確的處理函式。
* 為了加快處理速度，現代系統會使用一個**中斷向量表（interrupt vector）**：這是一張指向各裝置 ISR 的指標表格，通常儲存在低位址記憶體中（如前幾百個位址）。
* 每個中斷請求都附帶一個唯一編號，以作為查表的索引。

作業系統如 Windows 與 UNIX 都採用這樣的方式派發中斷。中斷架構還必須能夠保存被中斷任務的狀態資訊，以便處理結束後能正確還原。若中斷服務程式需要更動處理器狀態（例如改寫暫存器），它必須顯式地先保存當前狀態，再於結束時恢復。最後，中斷服務程式會將儲存的返回位址載入程式計數器中，並讓中斷前的程式繼續執行，就如同從未被中斷一樣。


### 1.2.1.2 實作（Implementation）

中斷的基本機制如下：CPU 硬體設有一條稱為\*\*中斷請求線（interrupt-request line）\*\*的電線，CPU 在執行每條指令後都會感測這條線是否有訊號。當某個控制器對中斷請求線發出訊號時，CPU 便會讀取中斷編號，並使用該編號作為索引，查詢中斷向量表中對應的中斷處理常式（interrupt-handler routine）地址。接著，CPU 會跳轉至該位址並開始執行該程式。

中斷處理程式首先儲存執行期間可能改變的狀態，接著判斷中斷原因，執行所需的處理動作，再還原狀態，最後執行「從中斷返回」的指令，讓 CPU 回到中斷前的狀態。整個過程中：

* 裝置控制器**觸發中斷**（assert interrupt）；
* CPU **捕捉中斷**（catches interrupt）；
* 中斷處理程式執行，並**清除中斷**（clear interrupt）來完成裝置服務；
* 如圖 1.4 所示，即為中斷驅動的 I/O 週期。

這套機制讓 CPU 能夠回應如裝置準備就緒這類**非同步事件**。然而，現代作業系統需要更高級的中斷處理能力，包括：

1. 能夠在關鍵處理期間**延遲中斷處理**；
2. 有效率地派發至**正確的中斷處理程式**；
3. 能辨別不同中斷**優先等級**，以依據緊急程度做出相應處理。

這三項功能由\*\*CPU 與中斷控制器硬體（interrupt-controller hardware）\*\*提供。

---

大多數 CPU 有兩條中斷請求線：

* **不可遮蔽中斷（nonmaskable interrupt）**：用於無法忽略的嚴重事件，如記憶體錯誤；
* **可遮蔽中斷（maskable interrupt）**：可在執行關鍵程式碼前由 CPU 暫時關閉，用於一般裝置請求服務。

中斷向量設計的目的，是避免讓單一中斷處理器遍歷所有可能的來源。在現實中，裝置數可能多於向量表所能容納的項目，因此常採用**中斷鏈結（interrupt chaining）**：向量表中每一項指向一串中斷處理器清單，逐一檢查直到找到能處理請求者為止。這種做法在處理效能與記憶體開銷之間取得平衡。

圖 1.5 顯示 Intel 處理器的中斷向量表，其中：

* 0～31 為**不可遮蔽中斷**，多用於錯誤通報；
* 32～255 為**可遮蔽中斷**，多由裝置觸發。

中斷機制也支援**中斷優先等級（interrupt priority levels）**，允許 CPU 延遲處理低優先級中斷，並允許高優先級中斷搶先執行。

---

總結來說，中斷機制在現代作業系統中無所不在，負責處理非同步事件與其他用途。無論是裝置控制器或硬體故障都會產生中斷，而中斷優先權機制則確保最緊急的工作能優先處理。由於中斷對時間敏感的處理至關重要，因此**高效的中斷處理能力**是系統效能的關鍵。


### 1.2.2 儲存結構（Storage Structure）

CPU 只能從記憶體載入指令，因此任何程式在執行前必須先被載入至記憶體。通用型電腦大多從**可寫入記憶體**執行程式，稱為**主記憶體（main memory）**，也就是**隨機存取記憶體（RAM）**。主記憶體多數是以\*\*動態隨機存取記憶體（DRAM）\*\*的半導體技術實作而成。

除了主記憶體外，電腦還使用其他類型的記憶體。例如，在電腦開機後最先執行的程式是**啟動程式（bootstrap program）**，負責載入作業系統。然而，RAM 是**揮發性（volatile）**的，會在斷電時喪失內容，因此無法儲存啟動程式。為此與其他目的，電腦會使用如**可電擦可程式唯讀記憶體（EEPROM）**與其他形式的**韌體（firmware）**，這些記憶體寫入次數極低，且具有非揮發性。

EEPROM 雖可重寫，但寫入速度慢，主要儲存靜態資料。例如 iPhone 就用它來儲存序號與硬體資訊。

---

所有記憶體皆以**位元組陣列**方式運作，每個位元組都有一個位址。資料互動透過「載入（load）」與「儲存（store）」指令進行，這些指令會從特定位址移動資料至 CPU 暫存器，或從暫存器寫入主記憶體。除了明確的載入/儲存外，CPU 也會自動依照程式計數器的位址從記憶體抓取指令執行。

在冯诺依曼架構中，一個典型的「取指→解碼→執行→儲存結果」的循環中，CPU 只是看到一連串的記憶體位址，並不知道這些地址是如何產生的（例如透過指令計數器、索引、間接存取等），也無法分辨這些地址是用於指令或資料。因此我們可以只關注這些執行時所產生的位址序列。

---

理想狀況下，所有程式與資料應該都常駐於主記憶體，但這在現實中不太可行，主要有兩個原因：

1. 主記憶體的容量通常不足以容納所有程式與資料；
2. 主記憶體為揮發性，斷電後內容會遺失。

因此大多數電腦系統會使用**次儲存裝置（secondary storage）**作為主記憶體的延伸，能夠**永久儲存大量資料**。

常見的次儲存裝置包括**硬碟（HDD）**與**非揮發性記憶體（NVM）**，例如快閃記憶體。大多數系統與應用程式在執行前皆儲存在次儲存中，並於執行時載入主記憶體。很多應用程式也會直接在次儲存中讀寫資料。然而，次儲存的速度遠低於主記憶體，因此其管理是系統設計的核心之一（詳見第11章）。

---

從更廣的角度看，我們所描述的記憶體架構（暫存器、主記憶體、次儲存）只是眾多架構之一，其他還包括**快取（cache）**、**光碟（CD-ROM/藍光）**、**磁帶**等。這些較慢、較大、通常只用於備份的裝置被稱為**第三層儲存（tertiary storage）**。

不同儲存系統的主要差異在於速度、容量與是否揮發。圖 1.6 顯示了根據容量與存取時間構成的**儲存階層**。原則上，記憶體越小越快，越接近 CPU。從圖中可見，越靠近上方的儲存系統速度越快，但容量越小。

---

由於儲存結構對作業系統設計至關重要，以下術語將貫穿全書使用：

* **揮發性儲存**：簡稱為「記憶體（memory）」，若需強調將會明確指定，如暫存器（register）。
* **非揮發性儲存（NVS）**：資料在斷電後仍保留。大多數情況下，我們所提到的 NVS 即為次儲存裝置。

非揮發性儲存又可分為兩類：

1. **機械式（Mechanical）**：如硬碟、光碟、磁帶、全息儲存等。若需強調某種裝置會另行指出。
2. **電子式（Electrical）**：如 flash、FRAM、NRAM、SSD。簡稱為 NVM。若需強調特定類型（例如 SSD）會明示。

通常機械式儲存容量較大且每位元成本較低，而電子式儲存則速度快但價格高、容量小。

一套完善的儲存系統設計，需兼顧上述各種要素：盡量減少昂貴記憶體的使用，並提供足夠且便宜的永久儲存容量。此外，在存取速度差異大的元件之間，安裝\*\*快取（cache）\*\*可顯著提升效能。


### 1.2.3 輸入/輸出結構（I/O Structure）

作業系統中有相當大一部分的程式碼專門負責管理**輸入/輸出（I/O）**，原因有二：

1. I/O 對系統的**穩定性與效能**至關重要；
2. I/O 裝置的性質多變，處理方式不一。

正如前面所提，一般用途的電腦系統由多種裝置組成，這些裝置皆透過\*\*共用匯流排（bus）\*\*進行資料傳輸。

---

在第 1.2.1 節所介紹的\*\*中斷驅動式 I/O（interrupt-driven I/O）\*\*雖適合小量資料傳輸，但對於大量資料（如非揮發性儲存）的傳輸，這種方式將導致大量中斷，產生過高的額外負擔。

為了解決這個問題，系統採用\*\*直接記憶體存取（DMA, Direct Memory Access）\*\*技術：

* 在 I/O 裝置進行傳輸前，系統會設定**緩衝區、指標、計數器**等資訊；
* 裝置控制器會將**整個區塊的資料**直接從裝置傳送至主記憶體或反向傳送，**完全不需要 CPU 介入**；
* 傳輸完成後只產生**一次中斷**通知，而非像低速裝置那樣每一個位元組都要產生一次中斷；
* 在傳輸期間，CPU 可以繼續處理其他工作，**大幅提升效率**。

---

某些高階系統使用**交換式（switch-based）架構**而非傳統的共用匯流排架構。在這類系統中，多個元件可以**同時彼此通信**，而非必須爭用匯流排的使用時機。在此架構下，DMA 的效能表現更加顯著。
